```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# n = int(input("Enter number of point signals: "))
x = np.array([5,6,7,8])
```

```python
def dft(x):
    n = len(x)
    X = np.zeros(n,dtype=complex)
    for i in range(n):
        for j in range(n):
            X[i] += x[j] * np.exp(-1j * 2 * np.pi * i * j/n)
    return X
```

```python
spectrum = dft(x)
```

```python
plt.subplot(2,1,1)
plt.stem(x)

plt.subplot(2,1,2)
plt.plot(spectrum)
```

```python
spectrum = dft(x1)
```

```python
plt.subplot(2,1,1)
plt.stem(x1)

plt.subplot(2,1,2)
plt.plot(spectrum)
```

```python
def alter(x):
    n = len(x)
    print(n)
    X = np.array([])
    for i in range(n):
        X = np.append(X,x[i])
        X = np.append(X,0)
    return X
```

```python
x = alter(x)
x
```

```python
spectrum = dft(x)
```

```python
plt.subplot(2,1,1)
plt.stem(x)

plt.subplot(2,1,2)
plt.plot(spectrum)
```

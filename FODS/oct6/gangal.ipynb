```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x[1:11],y,"*")
plt.ylabel("First Y")
plt.xlabel("First X")
plt.subplot(1,2,2)
plt.plot(y,x[1:11],"+")
plt.ylabel("2nd Y")
plt.xlabel("2nd X")
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=[15,15])
axes[0,0].set_title("UL")
axes[0,1].set_title("UR")
axes[1,0].set_title("LL")
axes[1,1].set_title("LR")
fig.tight_layout()
axes[0,0].plot(x,x,label = "Linear")
axes[0,1].plot(x,x**2,label = "Quad")
axes[1,0].plot(x,x**3, label = "Cubic")
axes[1,1].plot(x,x**4, label = "4th power")
axes[0,0].legend()
axes[0,1].legend()
axes[1,0].legend()
axes[1,1].legend()
```

```python
x = [1,2,3,4]
y = [11,22,33,44]
fig, ax = plt.subplots(figsize = (5,5))
ax.plot(x,y)

ax.set(title = 'sample',xlabel = 'x axis',ylabel = 'y axis')
ax.scatter(x,y)
```

```python
fig,ax = plt.subplots(nrows=3,ncols=2,figsize = (10,10))
ax[0,0].pie(df1['day'].value_counts(),labels = df1['day'].value_counts(), autopct='%1.2f',explode=[0.2,0,0,0])
ax[0,1].hist(x = df1['day'])
ax[1,0] = sns.regplot(data=df1,x='total_bill',y='tip')
ax[1,1].scatter(df1['total_bill'],df1['tip'],alpha=0.5,marker='d')
ax[2,0].hist(df1['total_bill'],bins=20)
```

```python
a = df1['total_bill']
mean = a.mean()
median = np.median(a)
mode = a.mode()
```

```python
sns.displot(a)
```

```python
sns.displot(a,kind='kde')
plt.axvline(mean, color = 'r', label = 'mean')
plt.axvline(median, color = 'g', label = 'median')
plt.axvline(mode[0], color = 'b', label = 'mode')
```

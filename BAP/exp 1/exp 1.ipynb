## Loading JSON file and converting to CSV

```python
import json
with open('./indian_states.json') as file:
    json_data = json.load(file)
```

```python
data = {"State ID":[state['state']['state_id'] for state in json_data],
        "State Name":[state['state']['state_name'] for state in json_data]}
```

```python
indian_states = pd.DataFrame(data)
indian_states.head()
```

```python
indian_states.to_csv('indian_states.csv')
```
